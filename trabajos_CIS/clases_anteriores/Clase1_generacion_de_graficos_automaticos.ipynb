* Funci칩n que genera en cada casion un a serie de puntos aleatorios
* Funcion que realiza las gr치ficas

```python
def valores():
    df=pd.DataFrame(
        {'x': range(1,21),
         'y': np.random.randn(20), 
         'z': np.random.randn(20)
         })
    return(df)

def dibujar(graf):
    plt.rcParams['figure.figsize'] = [7, len(graf)*2]# apra aumentar la iamgen den la visualizaci칩n de jupyter
    if len(graf) == 1:
        plt.plot(graf[0].x,graf[0].y)
    else:
        fig, axs = plt.subplots( len(graf), 1)
        for x in range(len(graf)):
            axs[x].grid(True);
            axs[x].plot(graf[x].x,graf[x].y);
            axs[x].set_ylabel('varaible en y{}'.format(x+1))  
    plt.xlabel('variable en x')  
    plt.show()
```

```python
graf=[valores(),valores(),valores(),valores(),valores()]
dibujar(graf)
```

## Otra forma para implementarlo

```python
def valores():
  df=pd.DataFrame(
    {'x': range(1,21),
     'y': np.random.randn(20), 
     'z': np.random.randn(20)
     })
  return(df)

def dibujar2(graf,filas,colum):
    plt.rcParams['figure.figsize'] = [colum*4, filas*2]# apra aumentar la iamgen den la visualizaci칩n de jupyter
    z=0;
    if filas*colum == len(graf):
        fig, axs = plt.subplots( filas, colum)
        for x in range(filas):
            for y in range(colum):
                axs[x,y].grid(True);
                axs[x,y].plot(graf[z].x,graf[z].y);
                axs[x,y].set_ylabel('varaible en y{}'.format(z+1)) 
                z+=1 
            axs[x,y].set_xlabel('variable en x')  
        plt.show()
    else:
        print('el tama침o no es el adecuado')
```

```python
graf=[valores(),valores(),valores(),valores(),valores(),valores()]
dibujar2(graf,3,2)
```
